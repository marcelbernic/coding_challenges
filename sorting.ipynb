```python
# Mergesort

a = [4, 7, 1, 9, 6, 5, 3, 8, 2]
print(a)

def merge(b, c, a):
    print("|---merging---|:")
    print("|->| b: ", b)
    print("|->| c: ", c)
    i, j, k = 0, 0, 0
    while i < len(b) and j < len(c):
        if b[i] <= c[j]:
            a[k] = b[i]
            i += 1
        else:
            a[k] = c[j]
            j += 1
        k += 1
    if i == len(b):
        a[k:len(b) + len(c)] = c[j:len(c)]
    else:
        a[k:len(b) + len(c)] = b[i:len(b)]
    print("|---result--->|", a)

def merge_sort(a):
    n = len(a)
    if n > 1:
        b = a[:len(a) // 2]
        c = a[len(a) // 2:]
        print("b: ", b)
        print("c: ", c)
        print("--------------------")
        merge_sort(b)
        merge_sort(c)
        merge(b, c, a)
        

merge_sort(a)
print(a)
```
