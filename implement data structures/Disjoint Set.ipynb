```python
import unittest


class TestDisjointSet(unittest.TestCase):

    def test_quick_find(self):
        uf = DisjointSetQuickFind(10)
        # 1-2-5-6-7    3-8-9    4
        uf.union(1, 2)
        uf.union(2, 5)
        uf.union(5, 6)
        uf.union(6, 7)
        uf.union(3, 8)
        uf.union(8, 9)
        self.assertTrue(uf.connected(1, 5))
        self.assertTrue(uf.connected(5, 7))
        self.assertFalse(uf.connected(4, 9))
        # 1-2-5-6-7 3-8-9-4
        uf.union(9, 4)
        self.assertTrue(uf.connected(4, 9))
        print('Success: test_quick_find')
        
    def test_quick_union(self):
        uf = DisjointSetQuickUnion(10)
        # 1-2-5-6-7    3-8-9    4
        uf.union(1, 2)
        uf.union(2, 5)
        uf.union(5, 6)
        uf.union(6, 7)
        uf.union(3, 8)
        uf.union(8, 9)
        self.assertTrue(uf.connected(1, 5))
        self.assertTrue(uf.connected(5, 7))
        self.assertFalse(uf.connected(4, 9))
        # 1-2-5-6-7 3-8-9-4
        uf.union(9, 4)
        self.assertTrue(uf.connected(4, 9))
        print('Success: test_quick_union')

def main():
    test = TestDisjointSet()
    test.test_quick_find()
    test.test_quick_union()
    
if __name__ == '__main__':
    main()
```

```python
# Quick Find Disjoint Set - store Root Vertex (instead of parent vertex)
# Construct -> O(N)
# Find      -> O(1)
# Union     -> O(N)
# Connected -> O(1)
class DisjointSetQuickFind(object):
    
    def __init__(self, size):
        self.root = [i for i in range(size)]
        
    def find(self, x):
        return self.root[x]
    
    def union(self, x, y):
        root_x = self.root[x]
        root_y = self.root[y]
        if root_x != root_y:
            for i in range(len(self.root)):
                if self.root[i] == root_y:
                    self.root[i] = root_x
    
    def connected(self, x, y):
        return self.find(x) == self.find(y)
    
```

```python
# Quick Union Disjoint Set - store Parent Vertex (instead of root vertex)
# Construct -> O(N)
# Find      -> O(H)
# Union     -> O(H)
# Connected -> O(H)
#    where H is the height of the tree
class DisjointSetQuickUnion(object):
    
    def __init__(self, size):
        self.root = [i for i in range(size)]
        
    def find(self, x):
        while x != self.root[x]:
            x = self.root[x]
        return x
    
    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            self.root[root_y] = root_x
    
    def connected(self, x, y):
        return self.find(x) == self.find(y)
    
```

```python
# Union by Rank Disjoint Set - store Parent Vertex (instead of root vertex)
# Construct -> O(N)
# Find      -> O(logN)
# Union     -> O(logN)
# Connected -> O(logN)
#    where H is the height of the tree
class DisjointSetUnionByRank(object):
    
    def __init__(self, size):
        self.root = [i for i in range(size)]
        self.rank = [1] * size
        
    def find(self, x):
        while x != self.root[x]:
            x = self.root[x]
        return x
    
    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)
        if root_x != root_y:
            if self.rank[root_x] > self.rank[root_y]:
                self.root[root_y] = root_x
            elif self.rank[root_x] < self.rank[root_y]:
                self.root[root_x] = root_y
            else:
                self.root[root_y] = root_x
                self.rank[root_x] += 1
    
    def connected(self, x, y):
        return self.find(x) == self.find(y)
```
